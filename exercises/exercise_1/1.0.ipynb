```python
import pandas as pd
import duckdb

df = pd.read_csv("data/1900-census-immigrant-pop-metro.csv")

df.head()
```

```python
columns = ["Sweden", "Norway", "Finland", "Denmark", "Total foreign born"]

df_nordic = df[columns]
df_nordic.head()
```

```python
df_nordic = df_nordic.fillna("0")
```

```python
for column in columns:
    df[column] = df_nordic[column].str.replace(",", "").astype(int)

df_nordic
```

```python
df_nordic = duckdb.query(
    """--sql
    WITH nordic AS (
        FROM df
            SELECT
                state,
                SUM("Total foreign born") AS "Total foreigners",
                SUM(sweden) AS Swedes,
                SUM(norway) AS Norweigans,
                SUM(denmark) AS Danish,
                SUM(finland) AS Finnish
            GROUP BY
                state
    )
    FROM nordic n
    SELECT *, n.swedes + n.norweigans + n.danish + n.finnish AS Nordic
    ORDER BY
        nordic
    DESC
""").df()

df_nordic
```

```python
df_nordic.shape
```

```python
import matplotlib.pyplot as plt

top_3_nordic = df_nordic.drop(["Total foreigners", "Nordic"], axis=1).set_index("State").iloc[:3]

top_3_nordic
```

```python
def nordic_plot():
    ax = top_3_nordic.plot(kind='bar', figsize=(12,6))
    ax.set_xticklabels(top_3_nordic.index, rotation = 0)

    fig = ax.get_figure()

    return fig, ax

nordic_plot()
```

```python
fig, ax = nordic_plot()

ax.containers
```

## Annotate

```python
fig, ax = nordic_plot()

arrowprops = dict(arrowstyle = "->", linewidth=2, connectionstyle = "arc3, rad=.3")
ax.annotate(text = "Swedes", xy=(0.8, 35000), arrowprops=arrowprops, xytext=(0.8, 42000))
```

```python
from matplotlib.ticker import FuncFormatter

def format(x, pos):
    if x >= 1000:
        return f'{x*1.0/1000:.0f}K'
    else:
        return f'{x:.0f}'


fig, ax = nordic_plot()

swede_bars, norweigan_bars, danish_bars, finnish_bars = ax.containers

def annotate_nordic(bar_artist, text, x_offset, y_offset, fontsize=12, fontweight="bold"):
    bar_x = bar_artist.get_x()
    bar_y = bar_artist.get_height()
    xy = (bar_x + 0.06, bar_y)
    xytext = (bar_x + x_offset, bar_y + y_offset)
    fontsize=fontsize
    fontweight=fontweight

    # Determine curve direction based on where the label is
    rad = 0.3 if xytext[0] > xy[0] else -0.3

    arrowprops = dict(
        arrowstyle="->",
        linewidth=2,
        connectionstyle=f"arc3,rad={rad}"
    )

    ax.annotate(
        text=text,
        xy=xy,
        xytext=xytext,
        arrowprops=arrowprops,
        fontsize=fontsize,
        fontweight=fontweight
    )

annotate_nordic(swede_bars[0], "57158", 0.4, -1000)
annotate_nordic(swede_bars[1], "34984", 0.4, 3000)
annotate_nordic(swede_bars[2], "29961", 0.4, 3000)


ax.set_title(
    "The Swedish population immigrated more to USA than the rest of the Nordic countries together during the 20th century",
    loc="left",
    pad=10,
    fontweight="bold"
    )

legend = ax.legend(
    facecolor='lightgrey',
    edgecolor='black',
    framealpha=0.8,
    frameon=True
)
ax.set_facecolor("lightgrey")
fig.patch.set_facecolor("lightgrey")

ax.set_xlabel("State", fontweight="bold", loc="left")
ax.set_ylabel("Amount of people", fontweight="bold", loc="top")

ax.xaxis.set_label_coords(0.03, -0.05)
ax.spines[["top", "right"]].set_visible(False)
ax.yaxis.set_major_formatter(FuncFormatter(format))


fig = ax.get_figure()
fig.tight_layout()
fig.savefig("figures/finnish_immigration_to_United_States.png")
```
