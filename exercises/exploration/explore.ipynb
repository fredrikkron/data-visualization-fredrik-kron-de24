```python
import duckdb

sold_cars_bar = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM modified_df
    GROUP BY 
        Year
    ORDER BY
        Year
    """
).df()

sold_cars_bar
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

sold_cars_modified = sold_cars_bar.iloc[:-1]

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x=sold_cars_modified["Year"], height=sold_cars_modified["Quantity"])
ax.set_title("Amount of cars sold between 2007 & 2016")
plt.xticks(rotation=45)
ax.set_xlabel("Year")
ax.set_ylabel("Quantity")
thousands_formatter = FuncFormatter(lambda val, pos: f"{int(val/1000)}K")
ax.yaxis.set_major_formatter(thousands_formatter)
```

```python
sold_cars_line = (
    duckdb.query(
        """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM modified_df
    GROUP BY 
        Year
    ORDER BY
        Year
    """
    )
    .df()
    .set_index("Year")
)

sold_cars_line
```

```python
from matplotlib.ticker import FuncFormatter

old_cars_modified = sold_cars_line.iloc[:-1].plot()
```

```python
CO2_per_year = (
    duckdb.query(
        """--sql
    SELECT
        Year,
        SUM(Avg_CO2) AS Average_CO2
    FROM modified_df
    GROUP BY 
        Year
    ORDER BY
        Year
    """
    )
    .df()
    .set_index("Year")
)

CO2_per_year
```

```python
ax2 = CO2_per_year.iloc[:-1].plot(
    legend=False,
    linewidth=4
    )

ax2.spines[["top", "right"]].set_visible(False)
ax2.set_title("Decrease in average CO2 emission since 2007", pad=20)
ax2.set_xlabel("Year", fontweight="bold")
ax2.set_ylabel("AVG CO2 Emission", fontweight="bold")
```
