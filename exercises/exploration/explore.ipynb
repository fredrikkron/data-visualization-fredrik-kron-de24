```python
import duckdb

sold_cars_bar = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM modified_df
    GROUP BY 
        Year
    ORDER BY
        Year
    """
).df()

sold_cars_bar
```

```python
import matplotlib.pyplot as plt

sold_cars_modified = sold_cars_bar.iloc[:-1]

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x=sold_cars_modified["Year"], height=sold_cars_modified["Quantity"])
ax.set_title("Amount of cars sold between 2007 & 2016")
plt.xticks(rotation=45)
ax.set_xlabel("Year")
ax.set_ylabel("Quantity")
```

```python
sold_cars_line = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM modified_df
    GROUP BY 
        Year
    ORDER BY
        Year
    """
).df().set_index("Year")

sold_cars_line
```
