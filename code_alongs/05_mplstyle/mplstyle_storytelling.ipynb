# mplstyle and DRY for storytelling in matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import duckdb

# in python script:
# Path(__file__).parent / "styles/base.mplstyle"
plt.style.use("styles/base.mplstyle")

df = pd.read_csv("data/jobs_in_data_2024.csv")
df.head()
```

```python
df.shape
```

```python
df_job_categories = df["job_category"].value_counts()
df_job_categories
```

```python
type(df_job_categories)
```

## visualizations

```python
ax = df_job_categories.plot(
    kind='barh',
    title="Number jobs by category for data jobs in 2024",
    xlabel="# OF JOBS",
    ylabel="JOB CATEGORY")

ax.invert_yaxis()
```
